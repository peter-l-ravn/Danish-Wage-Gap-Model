```python
import sympy as sp

c, At, mu_y, rho_h = sp.symbols('c A_t mu_y rho_h', real=True)
theta_hy, theta_ho, theta_ly = sp.symbols('theta_h_y theta_h_o theta_l_y', real=True)
Y_hh, Y_hl, Y_ll = sp.symbols('Y_hh Y_hl Y_ll', real=True)

coef_c = rho_h * (theta_ly * Y_ll - Y_hl * (theta_ho - theta_hy))

inner1 = (
    theta_hy**2 * Y_hh
    - (mu_y + 1) * theta_hy * theta_ly * Y_hl
    + mu_y * theta_ly**2 * Y_ll
)

inner2 = (
    (mu_y + 1) * theta_hy * theta_ly * Y_hl
    - theta_ho * mu_y * theta_ly * Y_hl
    - mu_y * theta_ly**2 * Y_ll
    + (theta_ho - theta_hy) * theta_hy * Y_hh
)

lhs = coef_c * (c - At * inner1)
rhs = (Y_hl * theta_hy + theta_ly * Y_ll) * rho_h * At * inner2

c_boundary = sp.solve(sp.Eq(lhs, rhs), c)[0]

print(sp.latex(c_boundary))
```
